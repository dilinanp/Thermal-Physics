# Specify the parameters for the two interacting Einstein Solids.

```python
# Number of oscillators in solid A
N_A = 300

# Number of oscillators in solid B
N_B = 200

# Total energy of the system (in units of h*f)
q = 100
```

# Generate the table.

```python
import numpy as np
import pandas as pd
from scipy.special import comb

df = pd.DataFrame(columns=['q_A', 'q_B', 'omega_A', 'omega_B', 'omega_total'])

df['q_A'] = np.arange(q+1)
df['q_B'] = q - df['q_A']

df['omega_A'] = comb(df['q_A']+N_A-1, df['q_A'])
df['omega_B'] = comb(df['q_B']+N_B-1, df['q_B'])
df['omega_total'] = df['omega_A']*df['omega_B']

omega_all = df['omega_total'].sum()

df['probability'] = df['omega_total']/omega_all

# Set display format for floating point numbers
pd.set_option('display.float_format', lambda x: f'{x:.0f}' if (x.is_integer() and x<100000) else f'{x:.2e}')

# Set pandas to display up to 120 rows
pd.set_option('display.max_rows', 120)

display(df)
print()
formatted_omega_all = f'{omega_all:.2e}' if omega_all >= 100000 else f'{omega_all:.0f}'
print(f'All accessible microstates = {formatted_omega_all}')
```

# Plot graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a figure with two subplots side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

label_fontsize = 14
tick_fontsize = 12
max_xticks = 11  # Desired maximum number of x-axis ticks

# Function to manually adjust x-ticks
def adjust_xticks(data, ax, max_xticks):
    # Ensure that the first and last labels are always included
    ticks = np.linspace(start=0, stop=len(data)-1, num=min(len(data), max_xticks), dtype=int)
    tick_labels = [int(data.iloc[tick]['q_A']) for tick in ticks]
    ax.set_xticks(ticks)
    ax.set_xticklabels(tick_labels)

# Plot omega_total vs q_A on the first subplot
sns.barplot(x='q_A', y='omega_total', data=df, ax=ax[0], color='royalblue')
sns.lineplot(x='q_A', y='omega_total', data=df, ax=ax[0],  marker='o', color='blue', lw=2)
ax[0].set_xlabel('$q_A$', fontsize=label_fontsize)
ax[0].set_ylabel('$\Omega_{total}$', fontsize=label_fontsize)
ax[0].set_title('$\Omega_{total}$ vs $q_A$', fontsize=label_fontsize)
adjust_xticks(df, ax[0], max_xticks)
ax[0].tick_params(axis='both', which='major', labelsize=tick_fontsize)

# Plot probability vs q_A on the second subplot
sns.barplot(x='q_A', y='probability', data=df, ax=ax[1], color='salmon')
sns.lineplot(x='q_A', y='probability', data=df, ax=ax[1],  marker='o', color='red', lw=2)
ax[1].set_xlabel('$q_A$', fontsize=label_fontsize)
ax[1].set_ylabel('$P(q_A)$', fontsize=label_fontsize)
ax[1].set_title('$P(q_A)$ vs $q_A$', fontsize=label_fontsize)
adjust_xticks(df, ax[1], max_xticks)
ax[1].tick_params(axis='both', which='major', labelsize=tick_fontsize)

# Adjust layout
fig.tight_layout()

# Show the plot
plt.show()
```
